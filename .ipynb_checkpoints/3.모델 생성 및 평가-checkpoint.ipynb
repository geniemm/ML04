```python
import pandas as pd
```

```python
df = pd.read_csv('data/loan_train_cleaning.csv')
df.head().T # 행렬바꿔주는거
```

### 머신러닝 변수
* label : 종속 변수 또는 정답
* feature : 독립 변수 또는 속성


```python
df.columns 
```

```python
# 정답(Loan_Status)을 도출하기위해 문제들(각 칼럼들 값)을 통한다
```

```python
features = ['Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area']
label = 'Loan_Status'
X,y = df[features], df[label]
```

```python
X.T # 정답은 없는 내용만 가져온다, 관례적으로 대문자 X는 독립변수를 넣어놓는다, 소문자 y는 정담에 관련된 변수
```

```python
y
```

```python
# 데이터셋을 구분해야하는데 이유는 알고리즘이 좋은건지 나쁜건지 판단해야하는데, 모든데이터셋을 알고리즘에 다 넣어버리면 
# 확인하기 어려우니까 8:2,7:3 정도의 비율로 쪼개놓는다.적은비율의 값으로 성능체크
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split # 자동으로 비율을 쪼개줌 (학습용_확인용_쪼갠다)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

### 모델 생성 및 학습
* 모델 생성 : 변수 = DecisionTreeClassifier()
* 학습 : 변수.fit(X_train, y_train)
    - 문제와 정답을 알고리즘에 학습
* 예측 : 변수.predict(X_test)
    - 확인용 문제를 predict에 넣으면 정답을 알려준다

대출이 되냐 안되냐는 분류문제(Classifier), 수치로 계산하는거는 레그레이션


```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train) # 학습하는명령어
```

```python
pred = dtc.predict(X_test) # 예측한 정답
pred
```

```python
y_test # 예측한정답이랑 실제정답이랑 다른데
```

```python
from sklearn.metrics import accuracy_score # 예측값과 실제값이 얼마나 동일한지 비교
```

```python
accuracy_score(y_test, pred) 
```
